```python
data_frame = sns.load_dataset("tips")
```

```python
data_frame.replace({"sex":{"Male":0, "Female":1}, "smoker":{"Yes":1, "No":0}}, inplace=True)
data_frame.head()
```

```python
days = pd.get_dummies(data_frame["day"])
data_frame = pd.concat([data_frame, days], axis=1)
data_frame.sample(2)
```

```python
times = pd.get_dummies(data_frame["time"])
data_frame = pd.concat([data_frame, times], axis=1)
# data_frame.drop(labels=["day", "Lunch", "Dinner"], inplace=True, axis=1)
data_frame.head()
```

```python
X = data_frame[["sex", "smoker", "size", "Thur", "Fri", "Sat", "Sun", "Lunch", "Dinner"]]
Y = data_frame[["tip"]]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=26)
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
predictions = regressor.predict(x_test)
```
